```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load Dataset
file_path = "../data/raw/Epileptic_Seizure_Recognition.csv"
df = pd.read_csv(file_path)
```

```python
# Step 1: Data Overview
print("First 5 Rows:")
df.head()
```

```python
print("Dataset Shape:", df.shape)
```

```python
print("Dataset Info:")
df.info()
```

```python
# Step 2: Checking Missing Values
missing_values = df.isnull().sum()
print("Missing Values in Dataset:")
missing_values[missing_values > 0]
```

```python
# Step 3: Checking Data Types
print("Data Types:")
df.dtypes.value_counts()
```

```python
# Step 4: Checking Target Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x=df['y'], palette='coolwarm')
plt.title('Target Variable Distribution')
plt.xlabel('Seizure Class')
plt.ylabel('Count')
plt.show()
```

```python
# Target variable distribution
df['y'].value_counts().plot(kind='bar', title='Class Distribution')
plt.show()
```

```python
# Step 5: Statistical Summary
print("Statistical Summary:")
df.describe()
```

```python
# Step 6: Checking Unique Values for Each Column
unique_values = df.nunique()
print("Unique Values per Column:")
unique_values
```

```python
# Step 7: Correlation Analysis
df_corr = df.drop(columns=['Unnamed'], errors='ignore').corr()
plt.figure(figsize=(12, 8))
sns.heatmap(df_corr, cmap='coolwarm', annot=False)
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
# Univariate Analysis
df.iloc[:, 1].hist()
plt.title('Feature1 Distribution')
plt.show()
```

```python
# Bivariate Analysis
sns.scatterplot(x=df.columns[1], y=df.columns[2], data=df)
plt.title('Feature1 vs Feature2')
plt.show()
```

```python
# Step 8: Pairplot of Some Features
selected_features = df.columns[1:6]  # First 5 features
sns.pairplot(df[selected_features])
plt.show()
```

```python
# Step 9: Distribution of Features
plt.figure(figsize=(15, 8))
for i in range(1, 11):  # First 10 numerical features
    plt.subplot(2, 5, i)
    sns.histplot(df.iloc[:, i], kde=True, bins=30)
    plt.title(df.columns[i])
plt.tight_layout()
plt.show()
```

```python
# Step 10: Checking Skewness of Features
numerical_columns = df.select_dtypes(include=['number']).columns  # Select only numerical columns
skewness = df[numerical_columns].skew()
print("Skewness of Features:")
skewness
```

```python
# Step 11: Checking Kurtosis of Features
numerical_columns = df.select_dtypes(include=['number']).columns  # Select only numerical columns
kurtosis = df[numerical_columns].kurt()
print("Kurtosis of Features:")
kurtosis
```

```python
# Step 12: Boxplot of Target Variable
df.boxplot(column=['y'])
plt.title('Boxplot of Target Variable')
plt.show()
```

```python
# Step 13: Feature Variance Analysis
numerical_columns = df.select_dtypes(include=['number']).columns  # Select only numerical columns
variances = df[numerical_columns].var()  # Calculate variance only for numerical columns

# Identify features with low variance
low_variance_features = variances[variances < 0.01].index.tolist()
print("Low Variance Features:", low_variance_features)
```

```python
# Step 14: Checking Class Imbalance
class_counts = df['y'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=class_counts.index, y=class_counts.values, palette='coolwarm')
plt.title('Class Imbalance Check')
plt.xlabel('Seizure Class')
plt.ylabel('Count')
plt.show()
```
